# Analysis of Flow control protocols

```python
import numpy as np
```

```python
def alpha(d,L,R):
    dv = d / (2 * (10**8))
    lr = R/L
    return dv * lr
```

```python
def stopwait(P,a):
    return (1-P)/(1+2*a)
```

```python
def gobackn(P,a,W):
    if(W>=(2*a)+1):
        return (1-P)/(1+(2*a*P))
    else:
        return (W*(1-P))/((2*a+1)*(1-P+W*P))
```

```python
def selectiveRepeat(P,a,W):
    if(W>=(2*a+1)):
        return 1-P
    else:
        return (W*(1-P))/(2*a+1)
```

# Packet size Vs Utilization

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, sharey=True)
fig.suptitle('Channel utilization VS Packet Size')
packets = np.arange(46,60, 2)
y_packet_stopwait = []
y_packet_goback = []
y_packet_selective = []
for p in packets:
    a = alpha(2000,p,10*(10**6))
    y_packet_stopwait.append(stopwait(0,a))
    y_packet_goback.append(gobackn(0,a,3))
    y_packet_selective.append(selectiveRepeat(0,a,3))
ax1.plot(packets, y_packet_stopwait,'ys-')
ax1.plot(packets, y_packet_goback,'rs-')
ax1.plot(packets, y_packet_selective,'bs-')
ax1.set_title("P = 0")
ax1.set_xlabel('Packet size')
ax1.set_ylabel('Channel Utilization')
ax1.legend(labels = ('Stop and Wait', 'Go back N','Selective Repeat'), loc = 'center right',prop={'size':18})

y_packet_stopwait = []
y_packet_goback = []
y_packet_selective = []
for p in packets:
    a = alpha(2000,p,5*(10**6))
    y_packet_stopwait.append(stopwait(0.25,a))
    y_packet_goback.append(gobackn(0.25,a,3))
    y_packet_selective.append(selectiveRepeat(0.25,a,3))
ax2.plot(packets, y_packet_stopwait,'ys-')
ax2.plot(packets, y_packet_goback,'rs-')
ax2.plot(packets, y_packet_selective,'bs-')
ax2.set_title("P = 0.25")
ax2.set_xlabel('Packet size')
ax2.set_ylabel('Channel Utilization')
ax2.legend(labels = ('Stop and Wait', 'Go back N','Selective Repeat'), loc = 'center right',prop={'size':18})

y_packet_stopwait = []
y_packet_goback = []
y_packet_selective = []
for p in packets:
    a = alpha(2000,p,5*(10**6))
    y_packet_stopwait.append(stopwait(0.5,a))
    y_packet_goback.append(gobackn(0.5,a,3))
    y_packet_selective.append(selectiveRepeat(0.5,a,3))
ax3.plot(packets, y_packet_stopwait,'ys-')
ax3.plot(packets, y_packet_goback,'rs-')
ax3.plot(packets, y_packet_selective,'bs-')
ax3.set_title("P = 0.5")
ax3.set_xlabel('Packet size')
ax3.set_ylabel('Channel Utilization')
ax3.legend(labels = ('Stop and Wait', 'Go back N','Selective Repeat'), loc = 'upper right',prop={'size':18})

fig = plt.gcf()
# for ax in fig.get_axes():
#     ax.label_outer()

fig.set_size_inches(18.5, 10.5)
fig.savefig('Packet.png', dpi=100)
```
